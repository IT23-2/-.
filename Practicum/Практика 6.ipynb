```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axisartist.axislines import *
fig = plt.figure(figsize = (5, 5), dpi = 500)
ax = fig.add_subplot(axes_class=AxesZero)

for direction in ["xzero", "yzero"]:
    # Добавляем ось x и y
    ax.axis[direction].set_visible(True)
for direction in ["left", "right", "bottom", "top"]:
    # Скрываем границы
    ax.axis[direction].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

def gen_rectangle(start, end):
    # Генерация прямоугольников
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    ax.scatter(x, y, marker='s', color='black')
    
def gen_triangle(start, end):
    # Генерация треугольников
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    ax.scatter(x, y, marker='^', color='red')

def gen_hexagon(start, end):
    # Генерация шестиугольников
    x = [i for i in range(start, end, 1)]
    y = [0] * (abs(start) + abs(end))
    ax.scatter(x, y, marker='H', color='orange')

# Генерация фигур
gen_rectangle(-3, 4)
gen_triangle(-3, 4)
gen_hexagon(-3, 4)
```
