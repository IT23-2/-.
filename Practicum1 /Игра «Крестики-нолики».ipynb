```python
print( " Игра Крестики-нолики для двух игроков ")

#создаем поля
board=list(range(1,10))

def fields(board):
    print('-' * 13)
    for i in range(3):
        print('|',board[0 + i * 3],'|',board[1 + i * 3],'|',board[2 + i * 3],'|')
        print('-'*13)

#проверка пользователя
def take_input(player_number1):
    valid=False
    while not valid:
        player_questions = input('Куда поставим' + player_number1 + '?')
        try:
            player_questions = int(player_questions)
        except:
            print('Hеправильно введены данные!')
            continue
        if player_questions >= 1 and player_questions <= 9:
            if(str(board[player_questions-1]) not in "XO"):
                board[player_questions-1] = player_number1
                valid = True
            else:
                print("Эта клетка уже занята!")
        else:
            print("Hеправильно введены данные. Введите число от 1 до 9.")
            
            
#проверка на выигрыша у пользователя
def check_win(board):
    win_move = ((0,1,2), 
                (3,4,5), 
                (6,7,8), 
                (0,3,6), 
                (1,4,7), 
                (2,5,8), 
                (0,4,8), 
                (2,4,6))
    for each in win_move:
        if board[each[0]] == board[each[1]] == board[each[2]]:
            return board[each[0]]
    return False

#
def main(board):
    counter = 0
    win = False
    while not win:
        fields(board)
        if counter % 2 == 0:
            take_input("X")
        else:
            take_input("O")
        counter += 1
        if counter > 4:
            cwb = check_win(board)
            if cwb:
                print(cwb, "Bыиграл!")
                win = True
                break
        if counter == 9:
            print("Ничья!")
            break
    fields(board)
main(board)

input("Нажмите Enter для выхода!")
```
