### Работа со строковыми данными в `pandas`

3.1  Добавьте в таблицу `recipes` столбец `description_length`, в котором хранится длина описания рецепта из столбца `description`.

```python
# Добавляем новый столбец 'description_length' в датафрейм recipes,
# содержащий длину описания каждого рецепта в символах
recipes = recipes.assign(description_length=list(len(str(recipes['description'][i])) for i in range(len(recipes)))

recipes
```

3.2 Измените название каждого рецепта в таблице `recipes` таким образом, чтобы каждое слово в названии начиналось с прописной буквы.

```python
# Проходим по первым 10 рецептам в датафрейме 
for i in range(10):
    # Получаем название рецепта
    q = recipes['name'][i]
    # Заменяем название рецепта на версию с заглавной первой буквой
    recipes = recipes.replace(q, q[0].upper() + q[1:])

# Выводим обновленный датафрейм recipes с названиями рецептов, у которых первая буква стала заглавной
recipes
```

3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
# Добавляем новый столбец 'name_word_count' в датафрейм ,
# содержащий количество слов в названии каждого рецепта
recipes = recipes.assign(name_word_count=list(len(list(filter(None, recipes['name'][i].split()))) for i in range(len(recipes))))

recipes
```

### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
# Преобразуем столбец 'contributor_id' в строку, объединяя его элементы через пробел
a = ' '.join(list(map(str, recipes['contributor_id'])))

# Создаем список уникальных строковых значений 'contributor_id' из датафрейма 
b = list(set(list(map(str, recipes['contributor_id']))))

# Создаем список, содержащий количество вхождений каждого уникального 'contributor_id' в строке a
c = list(a.count(b[i]) for i in range(len(b)))

# Создаем словарь contributor_recipes_counter, содержащий количество рецептов для каждого уникального 'contributor_id',
# сортируем его по количеству рецептов в обратном порядке
contributor_recipes_counter = dict(sorted({key: value for key, value in zip(b, c)}.items(), key=lambda x: x[1], reverse=True))
```

4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы? Обратите внимание, что отзыв с нулевым рейтингом или не заполненным текстовым описанием не считается отсутствующим.

```python
reviews
```

```python
# Создаем отсортированный список уникальных идентификаторов рецептов из датафрейма 
a = sorted(list(set(list(recipes['id']))))

# Создаем отсортированный список уникальных идентификаторов рецептов из датафрейма 
b = sorted(list(set(list(reviews['recipe_id']))))

# Инициализируем переменную k для подсчета рецептов без отзывов
k = 0

# Проверяем соответствие идентификаторов рецептов в обоих списках
for i in range(len(b)):
    if a[i] != b[i]:
        a.remove(a[i])
        k += 1

# Выводим количество рецептов, для которых отсутствуют отзывы
print(f'Отзывы отсутствуют для {k} рецептов')

# Создаем словарь average_rating, содержащий средний рейтинг для каждого рецепта
average_rating = {}
for i in b:
    average_rating[i] = reviews[reviews['recipe_id'] == i]['rating'].mean()

average_rating
```

4.3 Посчитайте количество рецептов с разбивкой по годам создания.

```python
import datetime as dt

# Добавляем новый столбец year в датафрейм recipes,
# содержащий год для каждого рецепта
recipes = recipes.assign(year=list(recipes['submitted'].dt.year))

# Создаем список уникальных годов из столбца year
a = sorted(list(set(list(recipes['year']))))

# Создаем словарь sort_by_year, содержащий количество рецептов для каждого года
sort_by_year = {}
for i in a:
    sort_by_year[i] = len(recipes[recipes['year'] == i])

sort_by_year
```
