```python
import pandas as pd
recipes = pd.read_csv('recipes_sample.csv')
reviews = pd.read_csv('reviews_sample.csv', index_col=0)
recipes = recipes.drop(columns=['contributor_id', 'n_steps'])
recipes
```

2. Случайным образом выберите 5% строк из каждой таблицы и сохраните две таблицы на разные листы в один файл `recipes.xlsx`. Дайте листам названия "Рецепты" и "Отзывы", соответствующие содержанию таблиц. 

```python
import numpy as np
reviews_index_list=sorted(list(set(list(reviews.index.values))))
list_len=len(reviews_index_list)
reviews_index_list=sorted([reviews_index_list[np.random.randint(list_len)] for i in range(int(list_len*0.05))])
reviews1=pd.DataFrame()
for j in list(reviews.columns.values):
    reviews1[j]=list(reviews.loc[reviews.index.values == i, j].iloc[0] for i in reviews_index_list)
reviews1
```

```python
recipes_index_list=sorted(list(set(list(recipes.index.values))))
list_len=len(recipes_index_list)
recipes_index_list=sorted([recipes_index_list[np.random.randint(list_len)] for i in range(int(list_len*0.05))])
recipes1=pd.DataFrame()
for j in list(recipes.columns.values):
    recipes1[j]=list(recipes.loc[recipes.index.values == i, j].iloc[0] for i in recipes_index_list)
recipes1
```

```python
import xlwings as xw
wb=xw.Book()
sht1 = wb.sheets.add(name = "Рецепты", after = wb.sheets[0].name)
sht2 = wb.sheets.add(name = "Отзывы", after = wb.sheets[1].name)
sht1.range('A1').value = recipes1
sht2.range('A1').value = reviews1
```

3. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_assign`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи присваивания массива значений диапазону ячеек.

```python
sht1.range('H1').options(ndim=1).value=['seconds_assign']
sht1.range('H2').options(transpose = True).value=list(map(lambda x: x*60, sht1.range('D2').expand('down').value))
```

4. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_formula`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи формул Excel.

```python
sht1.range('I1').options(ndim=1).value=['seconds_formula']
fml = xw.Range('I2').formula = f'=D2*60'
xw.Range('I2:I1501').formula = fml
```

5. Сделайте названия всех добавленных столбцов полужирными и выровняйте по центру ячейки.

```python
sht1['H1:I1'].font.bold = True
sht1.range('H1:I1').api.HorizontalAlignment = xw.constants.HAlign.xlHAlignCenter
```

6. Раскрасьте ячейки столбца `minutes` в соответствии со следующим правилом: если рецепт выполняется быстрее 5 минут, то цвет - зеленый; от 5 до 10 минут - жёлтый; и больше 10 - красный.

```python
sht1.activate()
a=sht1.range('D2:D1501').value
k=1
for i in a:
    k+=1
    if i<5:
        xw.Range((k, 4)).color = (0, 255, 0)
    elif 5<=i<=10:
        xw.Range((k, 4)).color = (255, 255, 0)
    else:
        xw.Range((k, 4)).color = (255, 0, 0)
```
