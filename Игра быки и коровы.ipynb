```python
print('Игра быки и коровы')

#Генерирует случайных чисел
from random import randint
a = randint(0,9)
b = randint(0,9)
c = randint(0,9)
d = randint(0,9)
num = str(a) + str(b) + str(c) + str(d)
attempt = True

##Ввод числа пользователем и корректность ввода
while attempt:
        player_num = input("Ваша догадка: ")
        buk = 0
        korova = 0
        for i in range(4):
            if num[i] == player_num[i]:
                buk += 1
            elif player_num[i] in num:
                korova += 1
        if buk == 4:
            print("4 быка! Вы выиграли!")
            attempt = False
        else:
            print(f'Количество быков - {buk}\nКоличество коров - {korova}\n')
```
