```python
print('-' * 26 + ' Клондайк'  + '-' * 26)
print('  |   ', end='')
for j in range(10):
    print(j, end = '     ')
print()

# Создаем поле размером 10x10
field = [[' ' for _ in range(10)] for _ in range(10)]

# Функция для отображения текущего состояния поля
def draw_field():
    for i,row in enumerate(field):
        print(i,'|',end = ' ')
        print('  |  '.join(row))
        print('-' * 61)

# Функция для проверки, идут ли три знака подряд
def check_win():
    for row in range(10):
        for col in range(10 - 2):
            # Проверка горизонтальных комбинаций
            if field[row][col] != ' ' and field[row][col] == field[row][col+1] == field[row][col+2]:
                return True
    for row in range(10 - 2):
        for col in range(10):
            # Проверка вертикальных комбинаций
            if field[row][col] != ' ' and field[row][col] == field[row+1][col] == field[row+2][col]:
                return True
    for row in range(10 - 2):
        for col in range(10 - 2):
            # Проверка диагональных комбинаций (слева-направо)
            if field[row][col] != ' ' and field[row][col] == field[row+1][col+1] == field[row+2][col+2]:
                return True
    for row in range(10 - 2):
        for col in range(2, 10):
            # Проверка диагональных комбинаций (справа-налево)
            if field[row][col] != ' ' and field[row][col] == field[row+1][col-1] == field[row+2][col-2]:
                return True
    return False

# Основная функция игры
def play_game():
    player = 1
    while True:
        draw_field()
        print(f"Ход игрока {player}")
        row = int(input("Введите номер строки (от 0 до 9): "))
        col = int(input("Введите номер столбца (от 0 до 9): "))
        
        # Проверяем, валидны ли введенные координаты
        if row < 0 or row >= 10 or col < 0 or col >= 10 or field[row][col] != ' ':
            print("Неверный ход. Попробуйте снова.")
            continue
        
        # Устанавливаем знак игрока на поле
        if player == 1:
            field[row][col] = 'x'
        elif player == 2:
            field[row][col] = 'x'

        # Проверяем, если текущий ход привел к выигрышу
        if check_win():
            draw_field()
            print(f"Игрок {player} проиграл!")
            break
        

        # Переключаемся на другого игрока
        if player == 1:
            player = 2
        else:
            player = 1
    
play_game()
```
