### Создайте двусвязный список для хранения информации о задачах в системе управления проектами. Каждый элемент списка должен содержать название задачи, описание, дату начала и дату окончания, а также список комментариев и прикрепленных файлов к задаче.

```python
# 35
class Node:
    def __init__(self, task_name, description, start_date, end_date):
        self.task_name = task_name
        self.description = description
        self.start_date = start_date
        self.end_date = end_date
        self.comments = []
        self.files = []
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def add_task(self, task_name, description, start_date, end_date):
        new_node = Node(task_name, description, start_date, end_date)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def delete_task(self, task_name):
        current = self.head
        while current:
            if current.task_name == task_name:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev
                return
            current = current.next

    def add_comment(self, task_name, comment):
        current = self.head
        while current:
            if current.task_name == task_name:
                current.comments.append(comment)
                return
            current = current.next

    def add_file(self, task_name, file):
        current = self.head
        while current:
            if current.task_name == task_name:
                current.files.append(file)
                return
            current = current.next

    def print_tasks(self):
        current = self.head
        while current:
            print(f"Task: {current.task_name}")
            print(f"Description: {current.description}")
            print(f"Start Date: {current.start_date}")
            print(f"End Date: {current.end_date}")
            print("Comments:")
            for comment in current.comments:
                print(comment)
            print("Files:")
            for file in current.files:
                print(file)
            print("------------")
            current = current.next


# Создание двусвязного списка
dll = DoubleLinkedList()

# Добавление задач
dll.add_task("Task 1", "Description 1", "2024-01-01", "2024-01-31")
dll.add_task("Task 2", "Description 2", "2024-02-01", "2024-02-28")
dll.add_task("Task 3", "Description 3", "2024-03-01", "2024-03-31")

# Добавление комментариев и файлов
dll.add_comment("Task 1", "Comment 1")
dll.add_comment("Task 1", "Comment 2")
dll.add_file("Task 1", "File 1.txt")
dll.add_file("Task 1", "File 2.txt")

# Вывод задач
dll.print_tasks()

# Удаление задачи
dll.delete_task("Task 2")

# Вывод задач после удаления
dll.print_tasks()
```
