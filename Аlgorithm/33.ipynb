```python
# 33
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, item):
        node = Node(item)
        node.next = self.top
        self.top = node

    def pop(self):
        if self.is_empty():
            return None
        item = self.top.data
        self.top = self.top.next
        return item

    def peek(self):
        if self.is_empty():
            return None
        return self.top.data

    def is_empty(self):
        return self.top is None

    def size(self):
        count = 0
        current = self.top
        while current:
            count += 1
            current = current.next
        return count


# Создание стека
stack = Stack()

# Добавление элементов
stack.push(1)
stack.push(2)
stack.push(3)

# Проверка размера стека
print(stack.size())  

# Удаление элементов, кратных 2
while not stack.is_empty():
    item = stack.pop()
    if item % 2 == 0:
        print(f"Удален элемент {item}")
    else:
        stack.push(item)

# Проверка размера стека
print(stack.size())  
```

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

    def is_empty(self):
        return len(self.stack) == 0

    def size(self):
        return len(self.stack)


# Создание стека
stack = Stack()

# Добавление элементов
stack.push(1)
stack.push(2)
stack.push(3)

# Проверка размера стека
print(stack.size())  

# Удаление элементов, кратных 2
while not stack.is_empty():
    item = stack.pop()
    if item % 2 == 0:
        print(f"Удален элемент {item}")
    else:
        stack.push(item)

# Проверка размера стека
print(stack.size())  
```
